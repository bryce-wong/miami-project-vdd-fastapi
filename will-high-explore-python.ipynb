## Load libraries

```python
import os
import urllib

import pandas as pd
```

## Load the data efficienty

```python
permit_data_filename = "miami_permit_data.csv"
```

```python
# Only download once
if not os.path.isfile(permit_data_filename):
    permit_data_url = "https://data.miamigov.com/api/views/2wxc-e53n/rows.csv?accessType=DOWNLOAD"
    urllib.request.urlretrieve(permit_data_url, permit_data_filename)
```

```python
permit_data_df = pd.read_csv(permit_data_filename)
with pd.option_context('display.max_columns', 999):
    display(permit_data_df)
```
